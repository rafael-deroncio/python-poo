```python
class Lutador:
    
    def __init__ (self, foto, nome, nacionalidade, altura, idade, peso, vitorias, derrotas, empates):
        self.__foto = foto
        self.__nome = nome
        self.__nacionalidade = nacionalidade
        self.__altura = altura
        self.__idade = idade
        self.__peso = peso
        self.__categoria = None
        self.__vitorias = vitorias
        self.__derrotas = derrotas
        self.__empates = empates
        
        self.defineCategoria(peso)

    @property
    def foto(self):
        return self.__foto

    @foto.setter
    def foto(self, foto):
        self.__foto = foto
    
    @property
    def nome(self):
        return self.__nome

    @nome.setter
    def nome(self, nome):
        self.__nome = nome

    @property
    def nacionalidade(self):
        return self.__nacionalidade

    @nacionalidade.setter
    def nacionalidade(self, nacionalidade):
        self.__nacionalidade = nacionalidade

    @property
    def altura(self):
        return self.__altura

    @altura.setter
    def altura(self, altura):
        self.__altura = altura

    @property
    def idade(self):
        return self.__idade

    @idade.setter
    def idade(self, idade):
        self.__idade = idade

    @property
    def peso(self):
        return self.__peso

    @peso.setter
    def peso(self, peso):
        self.__peso = peso

    @property
    def categoria(self):
        return self.__categoria

    @categoria.setter
    def categoria(self, categoria):
        
        self.__categoria = categoria
        
    @property
    def vitorias(self):
        return self.__vitorias

    @vitorias.setter
    def vitorias(self, vitorias):
        self.__vitorias = vitorias

    @property
    def derrotas(self):
        return self.__derrotas

    @derrotas.setter
    def derrotas(self, derrotas):
        self.__derrotas = derrotas

    @property
    def derrotas(self):
        return self.__derrotas

    @derrotas.setter
    def derrotas(self, derrotas):
        self.__derrotas = derrotas
        
    @property
    def empates(self):
        return self.__empates

    @empates.setter
    def empates(self, empates):
        self.__empates = empates
    
    def defineCategoria(self, peso):
        if peso < 52.2:
            self.categoria = 'Invalido.'
        elif peso <= 70.3:
            self.categoria = 'Leve'
        elif peso <= 83.9:
            self.categoria = 'Medio'
        elif peso <= 120.2:
            self.categoria = 'Pesado'
        else:
            self.categoria = 'Invalido'
        
    def apresentar(self):
        print(f'Lutador: {self.nome}')
        print(f'   Origem: {self.nacionalidade}')
        print(f'   {self.idade} Anos')
        print(f'   {self.altura}m de Altura')
        print(f'   Pesando: {self.peso} Kg')
        print(f'   Vitorias: {self.vitorias}')
        print(f'   Empates: {self.empates}')
        print(f'   Derrotas: {self.derrotas}')

    def status(self):
        print(self.foto, self.nome)
        print(f'   Peso {self.categoria}')
        print(f'   Vitorias: {self.vitorias}')
        print(f'   Empates: {self.empates}')
        print(f'   Derrotas: {self.derrotas}')

    def ganharLuta(self):
        self.vitorias += 1

    def perderLuta(self):
        self.derrotas += 1
    
    def empateLuta(self):
        self.empates += 1
```

```python
class Luta:
    
    def __init__(self, rounds):
        self.__desafiado = None
        self.__desafiante = None
        self.__rounds = rounds
        self.__aprovada = None
    
    @property
    def desafiado(self):
        return self.__desafiado

    @desafiado.setter
    def desafiado(self, lutador):
        self.desafiado = lutador

    @property
    def desafiante(self):
        return self.__desafiante

    @desafiante.setter
    def desafiante(self, lutador):
        self.desafiante = lutador

    @property
    def rounds(self):
        return self.__rounds

    @rounds.setter
    def rounds(self, rounds):
        self.rounds = rounds

    @desafiante.setter
    def desafiante(self, lutador):
        self.desafiante = lutador

    @property
    def aprovada(self):
        return self.__aprovada

    @aprovada.setter
    def aprovada(self, aprovada):
        self.aprovada = aprovada
    
    
    def marcarLuta(self, desafiado, desafiante):
        if desafiante.categoria == desafiado.categoria:
            if desafiante != desafiado:
                self.aprovada = True
                self.desafiante = desafiante
                self.desafiado = desafiado
            else:
                raise TypeError('desafiante != desafiado: Desafiante deve ser diferente do Desafiado!')
        else:
            raise TypeError('desafiante.categoria == desafiado.categoria: Categoria de Desafiante e Desafiado devem ser iguais!')

    
    def lutar(self):
        if self.aprovada:
            print(self.desafiado, self.desafiante)
        else:
            print('Luta NAO pode acontecer!')
```
